```python
import pandas as pd
```

```python
df = pd.read_csv('/content/transactions.csv')
df.head()
```

```python
df.drop('Transaction',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.fillna('',inplace=True)
```

```python
df.values.tolist()
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_ary = te.fit_transform(df.values.tolist())
```

```python
te_ary
```

```python
df_tran=pd.DataFrame(te_ary,columns=te.columns_)
```

```python
df_tran.head()
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
fi=apriori(df_tran,min_support=0.6,use_colnames=True)
```

```python
rules = association_rules(fi,min_threshold=.8)
```
